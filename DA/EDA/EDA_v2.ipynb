## Корреляционный анализ

Построим матрицу корреляции для выявления взаимосвязей между различными показателями.

```python
# Корреляционная матрица
correlation_matrix = df_segment.drop(['Дата/Время', 'Unnamed: 0'],axis = 1).corr()

# Визуализация корреляционной матрицы
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Корреляционная матрица')
plt.show()
```

С некоторыми показателями есть хорошая корреляция, возможно пригодится для модели предсказания.

## Выявление аномалий

```python
# Ящик с усами для выявления аномалий
plt.figure(figsize=(15, 8))
sns.boxplot(data=df_segment.drop(['Дата/Время', 'Unnamed: 0'],axis = 1))
plt.title('Выявление аномалий в токах для Афл3.1')
plt.ylabel('Ток (RMS)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

Видны те самые аномалии.

```python
# Ящик с усами для выявления аномалий
plt.figure(figsize=(15, 8))
sns.boxplot(data=df_segment[['Токовый ремни Афл3.1.Максим RMS', 'Токовый губки  Афл3.1.Максим RMS']])
plt.title('Выявление аномалий в токах для Афл3.1')
plt.ylabel('Ток (RMS)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
